## Exploratory Data Analysis (EDA)

### Explore Data day_df

Akan dilakukan eksplorasi data day_df untuk melihat rangkuman parameter statistik dari data day_df menggunakan method describe()

```python
day_df.describe(include="all")
```

Berdasarkan rangkuman parameter statistik di atas, diperoleh informasi data bahwa totak sepeda yang disewakan adalah 731 sepeda. Cuaca (weathersit) di data ini hanya ada 1 (nilai minimum) dan 3 (nilai maximum). Informasi ini akan membantu untuk menjawab pertanyaan keterkaitan cuaca dan penyewa sepeda.


Akan dilakukan demografi pelanggan berdasarkan cuaca (weathersit) dengan menggunakan method groupby() yang diikuti dengan method agg(). Berikut adalah kodenya.

```python
day_df.groupby(by="weathersit").agg({
    'dteday':'nunique',
    'cnt':'sum'
})
```

Berdasarkan pivot table di atas, dapat diketahui bahwa sepeda banyak digunakan oleh penyewa paling banyak pada cuaca 1 dan paling sedikit pada cuaca 3.

Akan dilakukan demografi pelanggan berdasarkan cuaca (weathersit) dengan menggunakan method groupby() yang diikuti dengan method agg() untuk melihat mean dan stdnya. Berikut adalah kodenya.

```python
day_df.groupby(by="weathersit").agg({
    "cnt": ["mean","std"]
})
```

Berdasarkan pivot table di atas, dapat diketahui bahwa jumlah rata-rata sepeda banyak digunakan oleh penyewa paling banyak pada cuaca 1 (Cerah, beberapa awan, sebagian berawan) , kemudian cuaca 2 (kabut berawan, kabut awan pecahm kabut beberapa awan dan kabut) dan paling sedikit pada cuaca 3 (hujan lebat yang ada es, badai petir dan kabut, juga salju berkabut.

### Explore Data hour_df

Akan dilakukan eksplorasi data hour_df untuk melihat rangkuman parameter statistik dari data hour_df menggunakan method describe()

```python
hour_df.describe(include="all")
```

Berdasarkan rangkuman parameter statistik di atas, diperoleh informasi data bahwa total sepeda yang disewakan adalah 731 sepeda. Jam (hr) penggunaan sepeda dimulai dari 0 dan 23 yang menunjukkan dalam 24 jam pasti ada yang menyewa sepeda. Informasi ini akan membantu untuk menjawab pertanyaan keterkaitan jam dan penyewa sepeda.

Akan dilakukan demografi pelanggan berdasarkan jam (hr) dengan menggunakan method groupby() yang diikuti dengan method agg(). Berikut adalah kodenya.

```python
hour_group = hour_df.groupby(by='hr').agg({
    'dteday':'nunique',
    'cnt':'sum'
})
```

Akan dilakukan pengurutan data untuk melihat jam yang paling banyak dan paling sedikit penyewaan sepeda menggunakan method sort_values.

```python
sorting_hour_group = hour_group.sort_values(by='cnt', ascending=False)
print(sorting_hour_group)
```

Berdasarkan hasil pengurutan data berdasarkan jam, diperoleh informasi bahwa penyewa sepeda terbanyak ada di jam 17.00 dengan jumlah 336860 pengguna dan penyewa sepeda terdikit ada di jam 04.00 dengan jumlah 4428 pengguna.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa cuaca yang paling banyak dan paling sedikit saat penggunaan sepeda?

Akan dibuat demografi mengidentifikasi jumlah penyewa berdasarkan cuaca (weathersit) menggunakan kode berikut.

```python
plt.figure(figsize=(21, 7))
ax = sns.barplot(x="weathersit", y="cnt", data=day_df)
ax.set_xlabel("Cuaca")
ax.set_ylabel("Jumlah penyewa")
ax.set_title("Jumlah Penyewa Sepeda Berdasarkan Cuaca", loc="center", fontsize=15)
plt.show()
```

Dari kode diatas, terlihat visualisasi data  yang menunjukkan bahwa cuaca nomor 1 memiliki paling banyak penyewa dan cuaca  nomor 3 memiliki paling sedikit penyewa.
